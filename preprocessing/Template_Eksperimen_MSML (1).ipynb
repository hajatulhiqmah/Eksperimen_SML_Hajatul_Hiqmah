```python
RAW_DATA_PATH = 'namadataset_raw/winequality-red.csv'

# Periksa apakah file ada
if not os.path.exists(RAW_DATA_PATH):
    print(f"File tidak ditemukan di {RAW_DATA_PATH}")
    print("Pastikan Anda sudah menjalankan perintah curl untuk mengunduh dataset,")
    print("atau unggah file 'winequality-red.csv' ke folder 'namadataset_raw'.")
else:
    # Dataset ini menggunakan titik koma (;) sebagai pemisah
    df = pd.read_csv(RAW_DATA_PATH, sep=';')
    
    # Tampilkan 5 baris pertama
    print("Data berhasil dimuat. Berikut 5 baris pertama:")
    print(df.head())
```

# **4. Exploratory Data Analysis (EDA)**

Pada tahap ini, Anda akan melakukan **Exploratory Data Analysis (EDA)** untuk memahami karakteristik dataset.

Tujuan dari EDA adalah untuk memperoleh wawasan awal yang mendalam mengenai data dan menentukan langkah selanjutnya dalam analisis atau pemodelan.

```python
if 'df' in locals():
    # 1. Menampilkan informasi dasar (tipe data, non-null counts)
    print("\n--- Info Dataset ---")
    print(df.info())

    # 2. Menampilkan statistik deskriptif
    print("\n--- Statistik Deskriptif ---")
    print(df.describe())

    # 3. Memeriksa nilai yang hilang (missing values)
    print("\n--- Cek Missing Values ---")
    print(df.isnull().sum())

    # 4. Visualisasi distribusi kolom target 'quality'
    print("\n--- Visualisasi Distribusi Kualitas (Asli) ---")
    plt.figure(figsize=(8, 5))
    sns.countplot(x='quality', data=df)
    plt.title('Distribusi Skor Kualitas Anggur (0-10)')
    plt.xlabel('Skor Kualitas')
    plt.ylabel('Jumlah')
    plt.show()

else:
    print("Harap jalankan sel 'Memuat Dataset' terlebih dahulu.")
```

# **5. Data Preprocessing**

Pada tahap ini, data preprocessing adalah langkah penting untuk memastikan kualitas data sebelum digunakan dalam model machine learning.

Jika Anda menggunakan data teks, data mentah sering kali mengandung nilai kosong, duplikasi, atau rentang nilai yang tidak konsisten, yang dapat memengaruhi kinerja model. Oleh karena itu, proses ini bertujuan untuk membersihkan dan mempersiapkan data agar analisis berjalan optimal.

Berikut adalah tahapan-tahapan yang bisa dilakukan, tetapi **tidak terbatas** pada:
1. Menghapus atau Menangani Data Kosong (Missing Values)
2. Menghapus Data Duplikat

5. Encoding Data Kategorikal
6. Binning (Pengelompokan Data)

Cukup sesuaikan dengan karakteristik data yang kamu gunakan yah. Khususnya ketika kami menggunakan data tidak terstruktur.

```python
#Type your code here

if 'df' in locals():
    print("Memulai preprocessing...")
    
    # 1. Membuat kolom target biner ('quality_category')
    # Kita akan mengubah ini menjadi masalah klasifikasi biner
    # Kualitas > 5 dianggap 'baik' (1), sisanya 'buruk' (0)
    df_processed = df.copy()
    df_processed['quality_category'] = df_processed['quality'].apply(lambda x: 1 if x > 5 else 0)

    # 2. Menghapus kolom 'quality' asli
    # Kolom ini tidak diperlukan lagi karena sudah diganti dengan 'quality_category'
    df_processed = df_processed.drop('quality', axis=1)

    print("Preprocessing selesai.")
    
    # Tampilkan hasil
    print("\n--- Data Setelah Preprocessing (5 baris pertama) ---")
    print(df_processed.head())

    # Cek distribusi target baru
    print("\n--- Distribusi Kategori Kualitas Baru (0=Buruk, 1=Baik) ---")
    print(df_processed['quality_category'].value_counts())
    
    # Visualisasi distribusi target baru
    plt.figure(figsize=(6, 4))
    sns.countplot(x='quality_category', data=df_processed)
    plt.title('Distribusi Kategori Kualitas (Biner)')
    plt.xlabel('Kategori (0 = Buruk, 1 = Baik)')
    plt.ylabel('Jumlah')
    plt.show()

    # Catatan: Kita akan menyimpan file ini di skrip 'automate_Nama-siswa.py'
    # untuk Kriteria 1 - Skilled.
    # Untuk sekarang, df_processed sudah siap di dalam memori.

else:
    print("Harap jalankan sel 'Memuat Dataset' terlebih dahulu.")
```
